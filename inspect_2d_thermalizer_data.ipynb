```python
import pickle
import numpy as np

with open("/scratch/pp3051/weather/main_repo/Data_2D_Thermalizer/output_data.pkl", "rb") as f:
    obj = pickle.load(f)

print(type(obj))
if isinstance(obj, dict):
    print(obj.keys())
    if "data" in obj:
        arr = obj["data"]
        print("data type:", type(arr))
        print("data shape:", np.asarray(arr).shape)
        print("data dtype:", np.asarray(arr).dtype)
    if "data_config" in obj:
        print("data_config:", obj["data_config"])
else:
    arr = np.asarray(obj)
    print("shape:", arr.shape, "dtype:", arr.dtype)
```
